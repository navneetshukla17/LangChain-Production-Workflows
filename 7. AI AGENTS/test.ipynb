```python
# Necessary imports
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage
from dotenv import load_dotenv

# Load api key
load_dotenv()
```

```python
# Model config
model = ChatGroq(model="llama-3.1-8b-instant")

# Tool
search_tool = DuckDuckGoSearchRun()

# Tool binding
model = model.bind_tools(tools=[search_tool])
```

```python
query = HumanMessage(content="Who is Elon musk ?")

model_result = model.invoke(query.content)

for tool_call in model_result.tool_calls:
    if tool_call["name"] == "duckduckgo_search":
        tool_result = search_tool.invoke(tool_call["args"])
        print(tool_result)
    
```
