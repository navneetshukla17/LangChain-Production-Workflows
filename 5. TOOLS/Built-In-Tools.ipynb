```python
from langchain_community.tools import DuckDuckGoSearchRun, DuckDuckGoSearchResults

query = "Who is Obama ?"

search_tool = DuckDuckGoSearchRun()
result1 = search_tool.invoke(query)

print(f"\nResults without source links:")
print(result1)

query = "Who is Obama ?"

search_tool = DuckDuckGoSearchResults()
result = search_tool.invoke(query)

# results with source links
print(f"\nResults with source links:")
print(result)
```

```python
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper

query = "Attack On Titan"

search_tool = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())
result = search_tool.invoke(query)

print(f"\nResults from Wikipedia:")
print(result)
```

```python
from langchain_experimental.utilities import PythonREPL

python_tool = PythonREPL()
result = python_tool.run("""
a = 10
b = 2

print(a * b)
                
""")
print(result)
```

```python
from langchain_community.tools import ShellTool

command = "rmdir test"

shell_tool = ShellTool()

result = shell_tool.invoke(command)
print(result)
```

```python
from langchain_community.tools import RequestsGetTool
from langchain_community.utilities import RequestsWrapper

query = "https://www.flipkart.com"

request_get_tool = RequestsGetTool(
    requests_wrapper=RequestsWrapper(),
    allow_dangerous_requests=True
)

result = request_get_tool.invoke(query)
print(result)
```

```python
from langchain_community.utilities import SQLDatabase
from sqlalchemy import create_engine

engine = create_engine(url="sqlite:///students.db")

sql_tool = SQLDatabase(engine=engine)

# Create table
sql_tool.run("""
CREATE TABLE student (
             id INT PRIMARY KEY,
             name TEXT,
             age INT
);
""")
print(f"\n========== Table created ==========")

# Insert records
sql_tool.run("""
INSERT INTO student (name, age)
VALUES ('Tushar', 23);
""")
print(f"\n========== Records inserted ==========")

# Retriever records
print(f"\n========== Retrieved records ==========")
sql_tool.run("""
SELECT * FROM student;
""")
```
