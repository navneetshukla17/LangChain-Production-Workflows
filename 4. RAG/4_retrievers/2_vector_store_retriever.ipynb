```python
from langchain_community.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.documents import Document
from dotenv import load_dotenv

# load huggingface api key
load_dotenv()
```

```python
# sample documents 
doc1 = Document(metadata={"squad": "marvel"}, page_content="Ironman is the main character in Marvel")
doc2 = Document(metadata={"squad": "marvel"}, page_content="Captain America is the most powerfull human character in Marvel")
doc3 = Document(metadata={"squad": "marvel"}, page_content="Thor is the GOD OF THUNDER")

docs = [doc1, doc2, doc3]
```

```python
embedding_model = HuggingFaceEmbeddings(model="sentence-transformers/all-MiniLM-L6-v2")
```

```python
# Chroma vector store
vector_store = Chroma.from_documents(
    documents=docs,
    embedding=embedding_model
)
```

```python
retriever = vector_store.as_retriever(search_kwargs={"k":2})
```

```python
query = "who is Ironman ?"
result = retriever.invoke(input=query)
```

```python
for i, doc in enumerate(result, start=1):
    print(f"\n----- Result:{i} -----\n")
    print(f"{doc.page_content}")
```
